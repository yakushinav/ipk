# Какие задачи решает машинное обучение?

Машинное обучение это один из инструментов искусственного интеллекта, который решает задачи связанные с исследованием изучением реального мира, поиском закономерностей и научных открытий, создания изобретений, автоматизации промышленного производства и повседневной жизни.

### Почему нужен новый способ решения задач?

Основной способ описания реальности в последние 300 лет это уравнения, в основном дифференциальные.

Дифференциальные уравнения описывают поведение некоторой системы в динамике, в зависимости от времени.

### Почему нужен новый способ решения задач? (Модель "хищник-жертва")

В природе система "хищник-жертва" взаимодействия популяций встречается достаточно часто. Например, в пруду обитают караси и щуки. В пруду достаточно питания карасям, а щуки питаются только карасями. По такой же системе взаимодействуют зайцы и волки, мыши и лисы и т. д.

$$
\begin{cases}{\frac{dx}{dt}}=(\alpha - \gamma y)x;\\{\frac{dy}{dt}}=(-\beta +\delta x)y.\end{cases}
$$

где $x$ — количество жертв, $y$ — количество хищников, $t$ — время, $\alpha ,\beta ,\gamma ,\delta$ — коэффициенты, отражающие взаимодействия между видами.

Мы перешли на новый уровень познания и аппарата дифференциальных уравнений уже не хватает для решения многих задач.

Почему? Потому что задачи стали сложные и зависят от огромного количества параметров.

## Линейная регрессия

Для предсказания меток класса по длине и ширине чашелистика, попробуем воспользоваться методом *линейная регрессия*:

\begin{equation*}
\tilde{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2.
\end{equation*}

-   $\tilde{y}$ — ответ модели
-   $x_1$ — длина чашелистика
-   $x_2$ — ширина чашелистика
-   $\theta_0, \theta_1, \theta_2$ — параметры модели (веса)

---

## Линейная регрессия (Проблема и решение)

Линейная регрессия выдает ответы в непрерывном диапазоне, а нам нужны ответы, принимающие значения -1 или 1.

Для решения этой проблемы воспользуемся функцией знака от ответов линейной регрессии:

$$
\tilde{y} = sign(\theta_0 + \theta_1 x_1 + \theta_2 x_2)
$$

$$
sign(t) =
 \begin{cases}
   1, &\text{t > 0}\\
   0, &\text{t = 0}\\
   -1, &\text{t < 0}
 \end{cases}
$$

---



````python
plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(data=df, x='sepal_length', y='sepal_width',
                         hue='species', palette='viridis', alpha=0.7)
k = 0.64
b = -0.2

x_min, x_max = df['sepal_length'].min(), df['sepal_length'].max()
x_values = np.linspace(x_min, x_max, 100)
y_values = k * x_values + b

plt.plot(x_values, y_values, color='red', linestyle='--', linewidth=3,
         label=f'y = {k}x + {b}')
````

Как видим, наши классы (виды цветка ириса) приблизительно можно разделить линией, значит имеет смысл построить такую разделяющую линию (или гиперплоскость, так как ее размерность на 1 меньше размерности пространства).

В общем виде это уравнение записывается вот так:
$$\theta_1x_1 + \theta_2x_2+b = 0$$

Или в более общем виде:
$$X\Theta = 0,$$
где $X=(x_1,\ldots,x_n)$ и $\Theta=(\theta_1,\ldots,\theta_n)$.

---

## Линейная регрессия (Интерпретация гиперплоскости)

Уравнение $X\Theta = 0$ задает гиперплоскость в пространстве признаков $X$ с вектором весов $\Theta$.

-   Если $X_i\Theta > 0$, то $i$-й объект попал по одну сторону от гиперплоскости
-   Если $X_i\Theta < 0$, то $i$-й объект попал по другую сторону от гиперплоскости.

А это значит, что наш *линейный классификатор*

\begin{equation*}
\tilde{y_i} = sign(X_i\Theta),
\end{equation*}

будет давать метку:
-   1, если объекты лежат по одну сторону от гиперплоскости
-   -1, если объекты лежать по другую сторону от гиперплоскости

Сам *линейный классификатор* строит **разделяющую гиперплоскость** в пространстве признаков.

---

## Функция ошибки

Для поиска весов модели $\Theta$, требуется задать функцию ошибки.
Для задачи классификации логичной функцией ошибки может служить **доля неправильных ответов** классификатора.

**Введем обозначения:**
-   $X$ — матрица признаков
-   $y$ — истинные метки
-   $X_i = \begin{pmatrix} 1& x_{i,1} & x_{i,2}\end{pmatrix}$ — признаки $i$-го объекта ($i$-го цветка ириса)
-   $y_i \in$ \{-1,1\} — истинная метка на $i$-м объекте
-   $\Theta = \begin{pmatrix}\theta_0 & \theta_1 & \theta_2\end{pmatrix}^T$

**Введем понятие отступа** на $i$-ом объекте: $M_i = y_i \cdot X_i\Theta$.
-   $M_i > 0$ — классификатор дал верный ответ
-   $M_i < 0$ — классификатор ошибся

Тогда *долю неправильных ответов* легко посчитать следующим образом:
$$
L = \dfrac{1}{N}\sum_{i=1}^N [M_i<0],
\begin{cases}
[True] = 1, \\
[False] = 0.
\end{cases}
$$

---

## Модель машинного обучения

Задача машинного обучения для построения модели линейного классификатора (машинное обучение с учителем):

Необходимо найти такие коэффициенты $\theta_1$ и $\theta_2$, чтобы обеспечить минимальное значение функции ошибки $L$.

**Модель машинного обучения:**
$\tilde{y}=sign(-0.4482+1.2155 * x1 -2.0299 * x2 )$

**Применим модель к нашим данным** (4 и 99 из таблицы на слайде 76):
-   $\tilde{y}=sign(-0.4482+1.2155 * 5.0 -2.0299 * 3.6)=sign(-1.67834)=-1$ => iris setosa
-   $\tilde{y}=sign(-0.4482+1.2155 * 5.9 -2.0299 * 3.0)=sign(0.63355)=1$ => iris virginica



````python
# Преобразуем классы в метки: setosa -> -1, virginica -> +1
df['label'] = df['species'].map({'Iris-setosa': -1, 'Iris-virginica': 1})

# Выделяем признаки и метки
X = df[['sepal_length', 'sepal_width']].values
y = df['label'].values

print(f"Размер данных: {X.shape}")
print(f"Классы: {np.unique(df['species'])}")
print(f"Метки: {np.unique(y)}")
````

````python
# Функция потерь (hinge loss)
def hinge_loss(y_true, y_pred):
    return np.maximum(0, 1 - y_true * y_pred)

# Функция для предсказания
def predict(X, theta):
    linear_output = theta[0] + np.dot(X, theta[1:])
    return np.sign(linear_output)

# Функция для вычисления accuracy
def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)
````

````python
# Градиентный спуск (пока можете считать, что это магия)
def gradient_descent(X, y, learning_rate=0.01, epochs=1000):
    n_samples, n_features = X.shape
    theta = np.zeros(n_features + 1)  # theta[0] - bias, theta[1:] - weights

    loss_history = []
    accuracy_history = []

    for epoch in range(epochs):
        # Прямое распространение
        linear_output = theta[0] + np.dot(X, theta[1:])
        predictions = np.sign(linear_output)

        # Вычисление потерь и accuracy
        loss = np.mean(hinge_loss(y, linear_output))
        acc = accuracy(y, predictions)
        loss_history.append(loss)
        accuracy_history.append(acc)

        # Вычисление градиента
        grad_theta0 = 0
        grad_theta = np.zeros(n_features)

        for i in range(n_samples):
            if y[i] * linear_output[i] < 1:
                grad_theta0 -= y[i]
                grad_theta -= y[i] * X[i]

        grad_theta0 /= n_samples
        grad_theta /= n_samples

        # Обновление весов
        theta[0] -= learning_rate * grad_theta0
        theta[1:] -= learning_rate * grad_theta

        if epoch % 100 == 0:
            print(f"Epoch {epoch}: Loss = {loss:.4f}, Accuracy = {acc:.4f}")

    return theta, loss_history, accuracy_history
````

````python
# Обучение модели
theta, loss_history, accuracy_history = gradient_descent(X, y, learning_rate=0.01, epochs=1000)

print(f"\nОбученные параметры:")
print(f"theta_0 (bias) = {theta[0]:.4f}")
print(f"theta_1 = {theta[1]:.4f}")
print(f"theta_2 = {theta[2]:.4f}")
````

````python
# Предсказания на обучающих данных
final_predictions = predict(X, theta)
final_accuracy = accuracy(y, final_predictions)
print(f"\nФинальная accuracy: {final_accuracy:.4f}")
````

````python
# Функция для классификации новых образцов
def classify_sample(sepal_length, sepal_width, theta):
    linear_output = theta[0] + theta[1] * sepal_length + theta[2] * sepal_width
    prediction = np.sign(linear_output)
    class_name = 'Iris-setosa' if prediction == -1 else 'Iris-virginica'
    confidence = abs(linear_output)  # чем дальше от границы, тем увереннее
    return class_name, confidence, linear_output

# Тестирование на новых данных
print("\nТестирование классификатора:")
test_samples = [
    [5.1, 3.5],  # Должен быть setosa
    [7.0, 3.2],  # Должен быть virginica
    [6.0, 2.8],  # Пограничный случай
    [5.5, 2.5]   # Пограничный случай
]

for i, sample in enumerate(test_samples):
    class_name, confidence, score = classify_sample(sample[0], sample[1], theta)
    print(f"Образец {i+1}: {sample} -> {class_name} (уверенность: {confidence:.3f}, score: {score:.3f})")

# Анализ разделимости
print(f"\nУравнение разделяющей прямой:")
print(f"{theta[1]:.4f} * x1 + {theta[2]:.4f} * x2 + {theta[0]:.4f} = 0")
print(f"Или: x2 = {(-theta[1]/theta[2]):.4f} * x1 + {(-theta[0]/theta[2]):.4f}")
````
